```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Load Dataset

```python

df = pd.read_csv("Titanic-Dataset.csv")
df.head()
```

## Basic Information

```python

df.info()
```

## Summary Statistics

```python

df.describe()
```

## Missing Values

```python

df.isnull().sum()
```

## Histograms for Numeric Features

```python
axes = df[['Age','Fare']].hist(bins=20, figsize=(10,4))

axes[0][0].set_title("Age Distribution")
axes[0][0].set_xlabel("Age")
axes[0][0].set_ylabel("Count")

axes[0][1].set_title("Fare Distribution")
axes[0][1].set_xlabel("Fare")
axes[0][1].set_ylabel("Count")

plt.tight_layout()
plt.show()
```

## Boxplots

```python
plt.figure(figsize=(8,4))

sns.boxplot(data=df[['Age','Fare']])

plt.title("Boxplot of Age and Fare")
plt.xlabel("Features")
plt.ylabel("Value")

plt.tight_layout()
plt.show()
```

## Survival Count

```python
sns.countplot(x='Survived', data=df)

plt.title("Survival Count of Passengers")
plt.xlabel("Survival Status (0 = Not Survived, 1 = Survived)")
plt.ylabel("Number of Passengers")

plt.tight_layout()
plt.show()
```

## Gender vs Survival

```python
sns.countplot(x='Sex', hue='Survived', data=df)

plt.title("Survival Count by Gender")
plt.xlabel("Gender")
plt.ylabel("Number of Passengers")
plt.legend(title="Survived", labels=["No", "Yes"])

plt.tight_layout()
plt.show()
```

## Correlation Matrix

```python
plt.figure(figsize=(10,6))

sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')

plt.title("Correlation Matrix of Titanic Dataset Features")
plt.tight_layout()
plt.show()
```

## Pairplot (Selected Features)

```python
pair = sns.pairplot(df[['Survived','Age','Fare','Pclass']].dropna(), hue='Survived')

pair.fig.suptitle("Pairplot of Key Features vs Survival", y=1.02)

plt.show()
```


## Observations

1. Females had higher survival rate.
2. Higher fare passengers survived more.
3. Age shows moderate impact.
4. Pclass is strongly correlated with survival.
